```python
# practice 4 
# calculate the mean incoming shortwave, outgoing shortwave, incoming longwave, and outgoing longwave radiation over the entire data set 
print(bsrn.SWU_Wm2.mean())
print(bsrn.SWD_Wm2.mean())
print(bsrn.LWD_Wm2.mean())
print(bsrn.LWU_Wm2.mean())
```

```python
#practice 5
#a. add a new column 'NET_SW' to bsrn with the net shortwave radiation
bsrn['NET_SW'] = bsrn['SWD_Wm2'] - bsrn['SWU_Wm2'] 
#b. add a new column 'NET_LW' to bsrn with the net longwave radiation
bsrn['NET_LW'] = bsrn['LWD_Wm2'] - bsrn['LWU_Wm2']
#c. add a new colymn net radiation to bsrn with net total radiation 
bsrn['NET_RAD'] = bsrn['SWD_Wm2'] - bsrn['SWU_Wm2'] + bsrn['LWD_Wm2'] - bsrn['LWU_Wm2']
```

```python
bsrn
```

```python
# data frame with daily mean values over the month of all these variables
# going to have to loop oover every day
#bsrn.index = date, pulling the days
daily_rad = []
#temporary variable is the first one possible , first day we will be getting the mean over that day. 
for d in bsrn.index.day.unique():
    avg_SWD = bsrn.SWD_Wm2[bsrn.index.day == d].mean()
    avg_SWU = bsrn.SWU_Wm2[bsrn.index.day == d].mean()
    avg_LWD = bsrn.LWD_Wm2[bsrn.index.day == d].mean() #add longwave downwelling
    avg_LWU = bsrn.LWU_Wm2[bsrn.index.day == d].mean() # add longwave 
    avg_NET_rad = bsrn.NET_RAD[bsrn.index.day == d].mean()
    daily_rad.append([d, avg_SWD, avg_SWU, avg_LWD, avg_LWU])
daily_RAD = pd.DataFrame(daily_rad, columns = ['avg_SWD','avg_SWU','avg_LWD','avg_LWU','avg_NET_rad'])
#appending the 
daily_RAD
```
