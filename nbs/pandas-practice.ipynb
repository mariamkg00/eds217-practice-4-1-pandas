Both series and dataframe objects contain an index object similar to the row index of the ndarray or the index of a list. 
pandas index object can be conceptualized as an immutable array or an ordered multiset. 


```python
import pandas as pd
import numpy as np
```

```python
#making series and dataframe objects from scratch; a series can be easily made from a list or array 
series = pd.Series([25.8, 16.2, 17.9, 18.8, 23.6, 29.9 ,23.6, 22.1])
series 
#the most common way to make a data frame is to use a list of lists or a dictionary 
#let's use a list of lists (or an array):
df= pd.DataFrame([[25.8, 28.1, 16.2, 11.0], [17.9, 14.2, 18.8, 28.0],
[23.6, 18.4, 29.9, 27.8],[23.6, 36.2, 22.1, 14.5]],
columns= ['A','B','C','D'])
df
```

```python
#making a dataframe from a dictionary
df = pd.DataFrame({'A': [25.8, 17.9, 23.6, 23.6],
                   'B': [28.1, 14.2, 18.4, 36.2],
                   'C': [16.2, 18.8, 29.9, 22.1],
                   'D': [11.0, 28.0, 27.8, 14.5]})
df

#using this method, each key corresponds to a column name(A,B,C,D), and each value is a column 
```

1. Make a new dataframe containing the first record for each day and the following columns : timestamp of the record, incoming shortwave radiation, direct and diffuse radiation, and incoming longwave radiation (BSRN collects data every minute)

```python
data_url = data_url = "https://raw.githubusercontent.com/environmental-data-science" \
            "/eds217_2023/main/data/BSRN_GOB_2019-10.csv"
bsrn= pd.read_csv(data_url)
```

```python

#stored as object, which means pandas cannot recognize the data type, it doesn't know how to handle them, but using the pd.to_datetime() 
#function can concert the column to datetime objects

bsrn.DATE = pd.to_datetime(bsrn.DATE)
#now that the column contains timestamps, we can set this column to be the index, using df.set_index()
bsrn.set_index('DATE',inplace= True)
bsrn
```
